```python
# Libraries
import math
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp
pd.set_option('max_columns', 50)

from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import expon
```

#### Merging the dataframes

```python
#your code here
order_items = pd.read_csv("olist_order_items_dataset.csv")
order_items
```

```python
order = pd.read_csv("olist_orders_dataset.csv")
order
```

```python
customers = pd.read_csv("olist_customers_dataset.csv")
customers
```

```python
order_cust = pd.merge(order, customers, on="customer_id")
order_cust
```

```python
order_cust_items = pd.merge(order_cust, order_items , on="order_id")
order_cust_items
```

```python
sellers = pd.read_csv("olist_sellers_dataset.csv")
sellers
```

```python
order_cust_sell = pd.merge(order_cust_items, sellers , on="seller_id")
order_cust_sell
```

```python
product_items = pd.read_csv("olist_products_dataset.csv")
product_items
```

```python
product_cust_sell = pd.merge(order_cust_items, product_items , on="product_id")
product_cust_sell
```

```python
geolocation = pd.read_csv("olist_geolocation_dataset.csv")
geolocation_1 = geolocation.rename(columns={"geolocation_zip_code_prefix": "customer_zip_code_prefix"})
geolocation_1
```
